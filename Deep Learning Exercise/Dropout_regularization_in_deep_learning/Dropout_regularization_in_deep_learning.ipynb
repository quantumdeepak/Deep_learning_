```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("sonar_dataset.csv",header=None)
df.sample(5)
```

```python
df.shape
```

```python
##  Check for nan values
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
x=df.drop(60,axis=1)
y=df[60]
y.head()
```

```python
y=pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.head()
```

```python
## Traning the model using deeplearning

## Model without droupout Layer
```

```python
model=keras.Sequential([
    ## 60 is the number of points in the intermodate layer
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')

])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

#### Model with some droupout layers

#### Model with speficic droupout rate allways need not to give consistent results allways and all thte times

######  the Droupout Regulerigation is mostely fallowed in imageprocessing where toomeny layers of nueral network are present


```python


model_d=keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    ## The droupout rate must be mentioned after each layer
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    ## We can vary the droupout rate
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model_d.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_d.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model_d.evaluate(x_test,y_test)
```

```python
y_pred=model_d.predict(x_test).reshape(-1)
print(y_pred[:10])
```

```python
# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test, y_pred))
```
